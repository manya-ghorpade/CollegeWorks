```python
# ============================================================
# RETRIEVAL AUGMENTED GENERATION (RAG) PIPELINE
# Using Hugging Face Transformers + FAISS + SentenceTransformers
# ============================================================

# ------------------------------------------------------------
# STEP 1: INSTALL REQUIRED LIBRARIES
# These libraries are required for building the RAG system
# ------------------------------------------------------------
# transformers -> provides pretrained language models
# datasets -> used to store and handle text datasets
# faiss-cpu -> efficient similarity search for retrieval
# sentence-transformers -> converts text into embeddings

!pip install transformers datasets faiss-cpu sentence-transformers
!pip install numpy

# ------------------------------------------------------------
# STEP 2: IMPORT REQUIRED LIBRARIES
# ------------------------------------------------------------

from datasets import Dataset                     # Used to create dataset of documents
from sentence_transformers import SentenceTransformer  # Used to generate text embeddings
import faiss                                     # Used for vector similarity search
import numpy as np                               # Used for numerical vector operations
from transformers import pipeline                # Used to load text generation model


# ------------------------------------------------------------
# STEP 3: CREATE A KNOWLEDGE BASE (DOCUMENT COLLECTION)
# In RAG, we need external knowledge that can be retrieved
# ------------------------------------------------------------

documents = [
    "Artificial Intelligence is the simulation of human intelligence in machines.",
    "Machine Learning is a subset of artificial intelligence that enables systems to learn from data.",
    "Deep Learning uses neural networks with multiple hidden layers to learn complex patterns.",
    "Natural Language Processing allows computers to understand and process human language.",
    "Transformers are deep learning models that are widely used in modern NLP systems."
]

# Convert the document list into a Hugging Face dataset
dataset = Dataset.from_dict({"text": documents})

print("Dataset Created:")
print(dataset)


# ------------------------------------------------------------
# STEP 4: LOAD AN EMBEDDING MODEL
# Embeddings convert text into numerical vectors so that
# similarity between documents can be computed
# ------------------------------------------------------------

embedding_model = SentenceTransformer('all-MiniLM-L6-v2')


# ------------------------------------------------------------
# STEP 5: CONVERT DOCUMENTS INTO VECTOR EMBEDDINGS
# Each document will be represented as a numerical vector
# ------------------------------------------------------------

doc_embeddings = embedding_model.encode(documents)

# Convert embeddings to numpy array for FAISS compatibility
doc_embeddings = np.array(doc_embeddings)

print("\nShape of Document Embeddings:")
print(doc_embeddings.shape)


# ------------------------------------------------------------
# STEP 6: CREATE FAISS INDEX
# FAISS stores embeddings and performs fast similarity search
# ------------------------------------------------------------

dimension = doc_embeddings.shape[1]   # dimension of embedding vectors

# Create FAISS index using L2 distance
index = faiss.IndexFlatL2(dimension)

# Add document embeddings into FAISS index
index.add(doc_embeddings)

print("\nNumber of Documents Indexed:", index.ntotal)


# ------------------------------------------------------------
# STEP 7: DEFINE DOCUMENT RETRIEVAL FUNCTION
# This function finds the most relevant documents for a query
# ------------------------------------------------------------

def retrieve_documents(query, k=2):

    # Convert query into embedding vector
    query_embedding = embedding_model.encode([query])

    # Perform similarity search in FAISS
    distances, indices = index.search(np.array(query_embedding), k)

    # Retrieve corresponding documents
    retrieved_docs = [documents[i] for i in indices[0]]

    return retrieved_docs


# ------------------------------------------------------------
# STEP 8: LOAD TEXT GENERATION MODEL
# This model generates answers based on retrieved context
# ------------------------------------------------------------

#generator = pipeline(
  #  "text-generation",     # task type
   # model="gpt2",          # pretrained generation model
   # max_length=100         # maximum generated tokens
#)

generator = pipeline(
    "text-generation",
    model="google/flan-t5-base",
    max_length=150,
    min_length=20
)

# ------------------------------------------------------------
# STEP 9: BUILD THE COMPLETE RAG PIPELINE
# This combines retrieval + generation
# ------------------------------------------------------------

def rag_pipeline(query):

    # ---- Retrieval Step ----
    # Retrieve relevant documents for the query
    retrieved_docs = retrieve_documents(query)

    # Combine retrieved documents into a context
    context = " ".join(retrieved_docs)

    # ---- Prompt Construction ----
    # Combine context and query into a prompt
    prompt = f"Context: {context}\n\nQuestion: {query}\nAnswer:"

    # ---- Generation Step ----
    # Generate answer using language model
    result = generator(prompt)[0]["generated_text"]

    return result


# ------------------------------------------------------------
# STEP 10: TEST THE RAG SYSTEM
# Provide a query and generate answer
# ------------------------------------------------------------

query = "What is machine learning?"

response = rag_pipeline(query)

print("\n===============================")
print("USER QUESTION:")
print(query)

print("\nGENERATED ANSWER:")
print(response)
print("===============================")
```
